```python
def w_k(k, N=64):
    return np.array([(exp((2 * 1j * pi * k * n) / N)) for n in range(64)])   
```

```python
def affiche_w_k(k):
    
    vec = w_k(k)
    reels = []
    imagins = []
    for x in vec:
        reels += [x.real]
        imagins += [x.imag]
    
    
    x1 = list(range(0, 64))
    
    plt.subplot(1, 2, 1)
    plt.stem(x1, reels)
    
    plt.subplot(1, 2, 2)
    plt.stem(x1, imagins)
    
    plt.show()

affiche_w_k(63)
    
```

```python
ks = [0, 1, 2, 3, 4, 16, 20, 30, 31, 32]

for k in ks:
    affiche_w_k(k)
    affiche_w_k(64-k)
```

```python
def scalaire(x, k):
    
    vec_k = w_k(k)
    return np.dot(x, vec_k)

x = [z for z in range(64)]
print(type(scalaire(x, 10)))
```
